```python
import numpy as np
import matplotlib.pyplot as plt

T = 1000
x = np.linspace(-2.5, 3.5, T)

f = 0.5 * x**4 - (x + 1)**3 - 3 * x**2
noise = np.random.normal(0, 6, T)
y = f + noise
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, f, label="Clean Polynomial Signal", linewidth=2, color='dodgerblue')
plt.plot(x, y, label="Noisy Signal", linewidth=1, alpha=0.6, color='orange')
plt.legend()
plt.title("Polynomial Signal with Changing Concavity", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
